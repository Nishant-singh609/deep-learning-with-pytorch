# this code represents the first step in building a neural network using PyTorch. It creates a simple neuron, takes input data, and produces an output based on weights and bias.
# In neural networks, the output is different every time because the weights and bias values are initialized randomly.

```python
import torch
import torch.nn as nn   # Used to build neural networks

# Create input tensor (multi-dimensional array)
x = torch.tensor([1.0, 2.0, 3.0])

# Create a single neuron (3 inputs → 1 output)
neuron = nn.Linear(in_features=3, out_features=1)

# Pass input through the neuron
output = neuron(x)

print(output)
```
